**1.- Lectura de archivos con "open"**

Se puede cargar la información de un archivo mediante la instrucción:

```python
File1 = open ("/ruta/del/archivo/ejemplo.txt","r")
```

Donde cargamos toda la información en una variable, especificando la ruta del archivo. 

Como segundo parámetro es el modo de acceso; "r" para "reading", "w" para "writting", o bien, "a" para "appending".

De esta forma se puede trabajar con el archivo de la siguiente manera:

```python
File1.name
```

Que regresará un string con el nombre del archivo, por ejemplo: "/ruta/nombre.txt"

En caso de que se quiera saber el modo en ql que se abrió el archivo, se puede acceder usando la intrucción: 

```python
File1.mode
```

Lo que regresará un string con una "r", "w" o una "a".


También se puede liberar el archivo con la instrucción:

```python
File1.close()
```

También se puede trabajar con una instrucción "with" para delimitar el tiempo que se tiene el archivo abierto como se muestra a continuación:

```python
with open("/ruta/archivo.txt") as Flie1:
  file_content = File1.read()
  print(file_content)
print(File1.closed)
print(file_content)
```

La ventaja de usar la instrucción "with" es que el archivo se cierra automáticamente al terminar con la indentación, sin embargo, aún podemos acceder al contenido de las variables que se crearon en la indentación de la instrucción "with".

En este ejmplo, el primer "print" después de la indentación imprimirá el si el archivo está cerrado, mientras que el segundo print seguirá imprimiendo el contenido del archivo.

Esto convertirá a la variable "file_content" en una lista. Cada renglón tendrá un "\n" al final, lo que indica el salto de renglón.

**2.- Escritura de un archivo con "open"**

También se puede crear archivos con la instrucción "open". Solo se tene que cambar el parámetro a "w":

```python
File1 = open("/ruta/archivo.txt", "w")
```

En caso de que ya exista el archivo en la ruta, este será sobreescrito.

De igual forma se puede usar la instrucción "with", pero hay que recordar que el archivo será cerrado una vez que la indentación termine. una vez que esto pase, ya no se podrá escribir más en el archivo. Un emplo de esto sería:

```python
with open("/ruta/archivo.txt", "w") as File1:
  File1.write("This is line A")
print("aquí ya no se puede escribir en el archivo")
```

Con la instrucción "wirte" podemos escribir en el archivo.

De igual forma que la lectura, se pude escribir una lista como renglones en un archivo:

```python
Lines = ["Renglon 1\n","Renglon 2\n","Renglon 3\n"]
with open("/ruta/archivo.txt","w") as File1:
  for line in Lines:
    File1.write(line)
```

En caso de que se quiera agregar contenido a un archivo existente, solo se tiene que abrir en modo "append" con el parámetro "a". De esta forma cuando usemos la instrucción "write", esta agregará las línes al final del archivo.
