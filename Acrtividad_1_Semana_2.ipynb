Con la instruccin "wirte" podemos escribir en el archivo.

De igual forma que la lectura, se pude escribir una lista como renglones en un archivo:

```python
Lines = ["Renglon 1\n","Renglon 2\n","Renglon 3\n"]
with open("/ruta/archivo.txt","w") as File1:
  for line in Lines:
    File1.write(line)
```
