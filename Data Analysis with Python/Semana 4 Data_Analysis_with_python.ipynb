Se importan las librerías que vamos a necesitar

```python
import pandas as pd
import numpy as np
```

En el siguiente código recuperamos la información y la almacenamos en una variable

```python
info = "https://raw.githubusercontent.com/PosgradoMNA/actividades-de-aprendizaje-HRaulVazquez/main/Data%20Analysis%20with%20Python/imports-85.data"
```

transformamos la variable creada en un dataframe con el que podemos trabajar mejor.

```python
df = pd.read_csv(info, header= None)

#Se asignan los nombres de las columnas
headers = ["Symboling","Normalized_Losses","Maker","Fuel-Type","Aspiration","No_of_Doors","Body_Style","Drive_wheels","Engine_location","Wheel_Base","Length","Width","Height","Curb_Weight","Engine_type","No_Of_Cilinders","Engine_Size","Fule_Type","Bore","Stroke","Compression_Ratio","Horsepower","Peak_Rpm","City_mpg","Highway_mpg","Price"]
df.columns = headers
df.head()
```

Vamos a ver la matriz descriptiva

```python
df.describe()
```

Sabemos que la información viene incompleta, se nesecita reemplazar todos los "?" por "NaN". Haremos esto guardando el resultado en un nuevo dataframe.

```python
NewDF = df.replace("?", np.NaN)
```

para fines del ejercicio, eliminaremos los rengloens que tengan el parámetro "Price" en null.

```python
df=NewDF.dropna(subset=["Price"], axis=0)
df.head()
```

Para obtener más información acerca d enuestro Dataframe se puede usar la siguiente instrucción:

```python
df.info()
```
