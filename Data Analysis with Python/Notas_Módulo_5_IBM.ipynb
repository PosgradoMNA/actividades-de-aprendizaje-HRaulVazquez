**Model Evaluation **

Nos sirve para evaluar el desempeño de nuestro modelo en el mundo real. Para esto podemos usar los datos recolectados.

Para esto podemos utilizar la instrucción de split en el conjunto de datos reunidos y dividirlos en un conjunto de datos para entrenamiento y para prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X_data,y_data, test_size 0.3, random_state = 0)
```

Donde "X_data" y "y_data" son nuestros conjuntos de datos; X son las variables independientes y "Y" es nuestra variable dependiente ( de salida o resultado )

Además, el test_size recibe un parámetro de 0 a 1. Donde los puntos decimales representan el porcentaje de datos a utilizar en el entrenamiento, en este caso el 30%.

El resutado de esta función son 4 grupos:
- X_train: variables independientes que se utilizarán para entrenamiento
- X_test: variables independientes que se utilizarán para probar el modelo
- Y_train: varables de salida para entrenamiento
- Y_test: variable de salida par probar el modelo

**Overfitting, Underfitting & Model Selection**

**Underfitting**: esto se presenta cuando el modelo ha sido configurado de tal forma que no logra predecir un porcentaje aceptable de datos.

**Overfitting**: Este caso se da cuando se configura el modelo de tal forma que el modelo se amolda lo suficiente como para predecir toda la información de entrenamiento, pero no logra predecir los datos de validación y su desempeño no es bueno en el mundo real.


Para seleccionar un buen modelo para el problema que se presente, se debe analizar el comportamiento de la información, así como la naturaleza del problema. Puede ser un problema de predicción o de clasificación, etc.

A fin de cuentas, se tiene que analizar el desempeño de más de un modelo y elegir el que mejor se adapte.

La ventaja de la regreción Ridge, es que esta previene el efecto de overfitting. Est lo logra con la correcta configuración del coeficiente "alpha"

Cuando se tiene elegido el modelo, pero se quiere buscar cual es la mejor configuración para este, la solución es utilizar la instrucción "**GridSearchCV**"

Esta función realiza una prueba con todos los parámetros que le proveamos para el modelo en cuestión. La manera en que le pasaremos todo el conjunto de parámetros es mediante un grid que contenga los parámetros de configuración.

Una vez que tenga todos los parámetros, este método organizará todas las combinaciónes posibles con estos parámetros y nos dará el resultado de los errores o de presioción, según lo hayamos configurado.
