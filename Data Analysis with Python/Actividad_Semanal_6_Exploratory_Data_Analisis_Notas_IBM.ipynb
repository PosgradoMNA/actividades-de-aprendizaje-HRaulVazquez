Analisis de Variantes (ANOVA)

Sirve para encontrar correlación entre diferentes grupos de variables categoricas. esta función arroja :
- F-Test score: Variación entre el promedio del grupo de muestra dividido entre la variación dentro del grupo
- P-value: Grado de confianza

Se puede asumir que existe una correlación fuerte entre una variable categórica y otra variable, si ANOVA nos devuelve un F-test grande y un P-value pequeño.

La correlación,
Es una medida estadística que indica la interdependencia entre dos variables. Por ejemplo, la lluvia y el no. de paraguas usados en una calle.

Si está lloviendo el no. de paraguas aumentará, mientras que si no está lloviendo, el no. de paráguas que se usan disminuirá seguramente.

Por tanto, podemos asumira que la lluvia y el no. de paraguas usados tiene una fuerte (alta) correlación. Pero también tenemos que hacer notar que "La correlación no denota causalidad" ya que, como en el ejemplo, no podemos saber si el no. de paraguas hizo que lloviera, o la lluvia hizo que el no. de paraguas aumentara.
